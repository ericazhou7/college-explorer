```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('disease_indicators.csv')
```

```python
df.columns
```

```python
df = df[(df['Question']=='Poverty') & (df['YearStart']==2016)]
```

```python
df = df[['LocationAbbr','DataValue','Stratification1','Latitude','Longitude']]
df = df.sort_values(by=['Stratification1','LocationAbbr'])
df
```

```python
g = df.groupby('LocationAbbr')
id_df = g['DataValue'].apply(lambda x: pd.Series(x.values)).unstack()
id_df.columns = g.agg({'Stratification1':lambda x:x.tolist()}).reset_index()['Stratification1'][0]
id_df
```

```python
id_df.to_csv('poverty_demographics.csv')
```
