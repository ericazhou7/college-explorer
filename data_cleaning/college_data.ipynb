```python
import numpy as np
import pandas as pd
```

```python
df2 = pd.read_csv('mrc_table2.csv')
df10 = pd.read_csv('mrc_table10.csv')
```

```python
df2_cols = ['super_opeid', 'female', 'par_mean', 'k_mean', 'mr_kq5_pq1']
df2_final = df2[df2_cols]

df10_cols = ['name','super_opeid','region','state','zip','tier','tier_name','type','iclevel','public','hbcu','flagship',
             'ipeds_enrollment_2013','sticker_price_2013','grad_rate_150_p_2013','sat_avg_2013','scorecard_netprice_2013',
             'scorecard_rej_rate_2013','scorecard_median_earnings_2011','asian_or_pacific_share_fall_2000',
             'black_share_fall_2000','hisp_share_fall_2000','alien_share_fall_2000','pct_arthuman_2000','pct_business_2000',
             'pct_health_2000','pct_multidisci_2000','pct_publicsocial_2000','pct_stem_2000','pct_socialscience_2000',
             'pct_tradepersonal_2000']
df10_final = df10[df10_cols]
```

```python
df = df2_final.join(df10_final.set_index('super_opeid'), on='super_opeid')
df.head()
```

```python
df2.shape
```

```python
df10.shape
```

```python
df.shape
```

```python
df['acc_rate_2013'] = 1 - df['scorecard_rej_rate_2013']
df.drop(['scorecard_rej_rate_2013'],axis=1)
```

```python
df['female_norm'] = (df['female']-df['female'].min())/(df['female'].max()-df['female'].min())
df['enrollment_norm'] = (df['ipeds_enrollment_2013']-df['ipeds_enrollment_2013'].min())/(df['ipeds_enrollment_2013'].max()-df['ipeds_enrollment_2013'].min())
df['price_norm'] = (df['sticker_price_2013']-df['sticker_price_2013'].min())/(df['sticker_price_2013'].max()-df['sticker_price_2013'].min())
df['grad_norm'] = (df['grad_rate_150_p_2013']-df['grad_rate_150_p_2013'].min())/(df['grad_rate_150_p_2013'].max()-df['grad_rate_150_p_2013'].min())
df['sat_norm'] = (df['sat_avg_2013']-df['sat_avg_2013'].min())/(df['sat_avg_2013'].max()-df['sat_avg_2013'].min())
df['netprice_norm'] = (df['scorecard_netprice_2013']-df['scorecard_netprice_2013'].min())/(df['scorecard_netprice_2013'].max()-df['scorecard_netprice_2013'].min())
df['acceptance_norm'] = (df['acc_rate_2013']-df['acc_rate_2013'].min())/(df['acc_rate_2013'].max()-df['acc_rate_2013'].min())
```
